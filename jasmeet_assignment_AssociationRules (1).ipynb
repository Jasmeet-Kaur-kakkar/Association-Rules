```python
data=pd.read_csv('book (1).csv')
data
```

```python
items=apriori(data,min_support=0.1,use_colnames=True)
items
```

```python
rules=association_rules(items,metric='confidence',min_threshold=0.5)
rules
```

```python
rules.sort_values('confidence',ascending=False)
```

```python
plt.scatter(data=rules,x='support',y='confidence')
```

```python
rules_lift=association_rules(items,metric='lift',min_threshold=0.8)
rules_lift
```

```python
rules_lift.sort_values('lift',ascending=False)
```

```python
plt.scatter(data=rules_lift,x='support',y='confidence')
```

```python
items1=apriori(data,min_support=0.04,use_colnames=True)
items1
```
